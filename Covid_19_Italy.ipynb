```python
#Basic Analysis of Covid 19 Italy data in order to draw conclusion to governing authority

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("covid19_italy_region.csv")
```

```python
data.head(10)
```

```python
data.columns
```

```python
%matplotlib inline
data.describe()
```

```python
#recovery rate
sns.relplot(x='TotalPositiveCases',y='Recovered',data=data)
```

```python
#death rate with respect to total cases
sns.relplot(x='TotalPositiveCases',y='Deaths',data=data)
```

```python
#Decrease in new positive cases with increase in test numbers
sns.relplot(x='TestsPerformed',y='NewPositiveCases',data=data)
```

```python
data.columns
```

```python
#Intensive care increased with increased number of cases
sns.relplot(x='TotalHospitalizedPatients',y='HospitalizedPatients',hue='IntensiveCarePatients',data=data)
```
